```python
def matint(x):     # Turn a raw matrix string from Project Euler to a 2 layer list with int elements
    x=x.split()
    for i in range(len(x)):
        x[i]=strist(x[i])
    return x

def listint(x):
    out=''.join(x)
    return int(out)

def strint(x): # Turns string to integer for all items in list
    for i in range(len(x)):
        x[i]=int(x[i])
    return x

def strist(x): # Turns string to list
    nist=[]
    for i in x:
        nist.append(i)
    return nist

def intstr(x):  # Turns integer to string for all items in list
    for i in range(len(x)):
        x[i]=str(x[i])
    return x
```

```python
def choice(pos):       # possible choice for position (y,x) of grid
    global grid
    y=pos[0]
    x=pos[1]
    num = {str(i) for i in range(1,10)}
    ytaken = set(grid[y])-{'0'}
    xtaken = set(i[x] for i in grid)-{'0'}
    num = num - ytaken - xtaken - sub(pos)
    return sorted(list(num))

def sub(pos):    # which sub grid does pos belongs to?  # numbers taken in sub grid 
    global grid
    y=pos[0]
    x=pos[1]
    y = y-y%3
    x = x-x%3
    return set(grid[y][x:x+3] + grid[y+1][x:x+3] + grid[y+2][x:x+3])-{'0'}

def allchoice():       # make a dictionary of possible choices for all cells in grid based on basic inspection
    global grid, dick, c
    dick = dict()
    for i in range(9):
        for j in range(9):
            pos= (j,i)
            if grid[j][i]=='0':
                dick[pos] = choice(pos)
    for i in dick:                 # if see single choice in a cell, fill it in
        if len(dick.get(i))==1:
            grid[i[0]][i[1]]=dick.get(i)[0]
```

```python
def subdick(sub):       # check unique choices in a subgrid and fills it in the grid once it finds one
    global dick
    global grid
    num = [str(i) for i in range(1,10)]
    rist = relsub(sub)
    #print(f'subgrid {sub}')
    for i in num:
        s=0
        for pos in rist:
            if dick.get(pos)!=None:
                if i in dick.get(pos):
                    s+=1
        #print(i,s)
        if s==1:
            for pos in rist:
                if dick.get(pos)!=None:
                    if i in dick.get(pos):
                        #print('unique:',i,(pos))
                        grid[pos[0]][pos[1]]=i
                        update2(pos, i)
    update()
```

```python
def rowdick(row):       # check unique choices in a subgrid and fills it in the grid once it finds one
    global dick
    global grid
    num = [str(i) for i in range(1,10)]
    rist = relrow(row)
    #print(f'row {row}')
    for i in num:
        s=0
        for pos in rist:
            if dick.get(pos)!=None:
                if i in dick.get(pos):
                    s+=1
        #print(i,s)
        if s==1:
            for pos in rist:
                if dick.get(pos)!=None:
                    if i in dick.get(pos):
                        #print('unique:',i,(pos))
                        grid[pos[0]][pos[1]]=i
                        update2(pos, i)
    update()
```

```python
def coldick(col):       # check unique choices in a subgrid and fills it in the grid once it finds one
    global dick
    global grid
    num = [str(i) for i in range(1,10)]
    rist = relcol(col)
    #print(f'row {col}')
    for i in num:
        s=0
        for pos in rist:
            if dick.get(pos)!=None:
                if i in dick.get(pos):
                    s+=1
        #print(i,s)
        if s==1:
            for pos in rist:
                if dick.get(pos)!=None:
                    if i in dick.get(pos):
                        #print('unique:',i,(pos))
                        grid[pos[0]][pos[1]]=i
                        update2(pos, i)
    update()
```

```python
def rel(pos):    # returns positions related to the input position (same row, same column, same subgrid)
    y=pos[0]
    x=pos[1]
    rist=[]
    for j in range(9):
        rist.append((j,x))
    for i in range(9):
        rist.append((y,i))
    y0 = y-y%3
    x0 = x-x%3
    for i in range(x0,x0+3):
        rist.append((y0,i))
        rist.append((y0+1,i))
        rist.append((y0+2,i))
    return set(rist)
```

```python
def relsub(sub):    # returns positions related to the input subgrid (which is indexed normally)
    rict={1:(0,0), 2:(0,3), 3:(0,6), 4:(3,0), 5:(3,3),6:(3,6), 7:(6,0), 8:(6,3), 9:(6,6)}
    pos=rict.get(sub)
    y=pos[0]
    x=pos[1]
    rist=[]
    for i in range(x,x+3):
        rist.append((y,i))
        rist.append((y+1,i))
        rist.append((y+2,i))
    return rist
```

```python
def relrow(y):    # returns positions related to the input row (which is indexed starting with 0)
    rist=[]
    for i in range(9):
        rist.append((y,i))
    return rist
```

```python
def relcol(x):    # returns positions related to the input row (which is indexed starting with 0)
    rist=[]
    for j in range(9):
        rist.append((j,x))
    return rist
```

```python
def unique(x):      # strategy 2: find unique choices, x = cycles
    global grid
    global dick
    update()
    for i in range(x):
        for j in range(1,10):     # sub 2 p
            subdick(j)
        for j in range(9):
            rowdick(j)
        for j in range(9):
            coldick(j)  
```

```python
def delete(pos ,num):        # delete dick choice (num) of  pos
    global dick
    if dick.get(pos)!=None:
        if num in dick.get(pos):
            dick.get(pos).remove(num)
```

```python
def straight(pist):    # check if all positions in pist is on a striaght line
    sy=set(i[0] for i in pist)         # horizontal
    sx=set(i[1] for i in pist)         # vertical
    if len(sy)==1:
        return relrow(list(sy)[0])        # = row
    if len(sx)==1:
        return relcol(list(sx)[0])       #  = col
    return 0
```

```python
def lock(sub):       # technique 3: locked candidates,  criteria: pos s with same choice in straight line in a subgrid
    global dick
    global grid
    pist=[]
    num = [str(i) for i in range(1,10)]
    rist = relsub(sub)
    #print(f'subgrid{sub}')
    for i in num:
        s=0
        for pos in rist:
            if dick.get(pos)!=None:
                if i in dick.get(pos):
                    s+=1
        #print(i,s)    
        if s==2 or s==3:
            pist=[]
            for pos in rist:
                if dick.get(pos)!=None:
                    if i in dick.get(pos):
                        pist.append(pos)
            if straight(pist)!=0:
                #print(pist)
                chist=(set(straight(pist)) ^ set(relsub(sub)))-set(relsub(sub))    # positions to be updated in dick
                for pos in chist:
                    delete(pos, i)
                
```

```python
def collock(col):       # technique 3: locked candidates,  criteria: pos s with same choice in a subgrid in the same colomn
    global dick
    global grid
    pist=[]
    num = [str(i) for i in range(1,10)]
    rist = relcol(col)
    for i in num:
        s=0
        for pos in rist:
            if dick.get(pos)!=None:
                if i in dick.get(pos):
                    s+=1
        #print(i,s)    
        if s==2 or s==3:
            pist=[]                 # pist is list of locked in pos 
            for pos in rist:
                if dick.get(pos)!=None:
                    if i in dick.get(pos):
                        pist.append(pos)
            
            if insub(pist)!=0:
                #print(pist)
                chist=(set(insub(pist)) ^ set(relcol(col)))-set(relcol(col))    # positions to be updated in dick
                for pos in chist:
                    delete(pos, i)
```

```python
def rowlock(row):       # technique 3: locked candidates,  criteria: pos s with same choice in a subgrid in the same row
    global dick
    global grid
    pist=[]
    num = [str(i) for i in range(1,10)]
    rist = relrow(row)
    for i in num:
        s=0
        for pos in rist:
            if dick.get(pos)!=None:
                if i in dick.get(pos):
                    s+=1
        #print(i,s)    
        if s==2 or s==3:
            pist=[]                 # pist is list of locked in pos 
            for pos in rist:
                if dick.get(pos)!=None:
                    if i in dick.get(pos):
                        pist.append(pos)
            
            if insub(pist)!=0:
                #print(pist)
                chist=(set(insub(pist)) ^ set(relrow(row)))-set(relrow(row))    # positions to be updated in dick
                for pos in chist:
                    delete(pos, i)
```

```python
def allock(x):
    global grid
    global dick
    for i in range(x):
        for j in range(1,10):
            lock(j)
        for j in range(9):
            collock(j)
            rowlock(j)
```

```python
def old_relsub(pos):    # returns positions related to the input position (same subgrid)
    y=pos[0]
    x=pos[1]
    rist=[]
    for i in range(x,x+3):
        rist.append((y,i))
        rist.append((y+1,i))
        rist.append((y+2,i))
    return set(rist)
```

```python
def insub(pist):           # check if the list of pos in pist are in the same subgrid
    rist=[]
    for i in pist:
        y=i[0]
        x=i[1]
        pos0=(y-y%3,x-x%3)
        rist.append(pos0)
    if len(set(rist))==1:
        return old_relsub(rist[0])
    return 0
```

```python
def dou():       # technique 4: naked doubles,  criteria: pos s with two same choices in the same row
    global dick
    global grid
    cist=[]       # list of two_choices
    dou_cist=[]     # list of naked doubles
    pist=[]           # list of pos with same two choices
    choice_double=()
    num = [str(i) for i in range(1,10)]
    for pos in dick:
        if len(dick.get(pos))==2:
            cist.append(tuple(dick.get(pos)))
    for i in set(cist):
        if cist.count(i)>=2:
            dou_cist.append(i)    
    
    for i in dou_cist:
        pist=[]
        for pos in dick:
            if dick.get(pos)==list(i):
                pist.append(pos)
        #print(i,'location:',pist)
        for j in nc2(pist):                 # check if any two pos in pist is in a straight line
            if straight(j)!=0:

                chist= set(straight(j))-set(j)
                #print(chist)
                for pos in chist:
                    delete(pos, i[0])
                    delete(pos, i[1])
            if insub(j)!=0:
                chist= set(insub(j))-set(j)
                #print(chist)
                for pos in chist:
                    delete(pos, i[0])
                    delete(pos, i[1])
```

```python
def update():        # if see single choice in a cell, fill it in and update choices of related positions
    global dick
    global grid
    single_choice=''
    rist=[]
    for i in dick.copy():                 
        if len(dick.get(i))==1:
            single_choice = dick.get(i)[0]
            grid[i[0]][i[1]] = single_choice
            for j in rel(i):
                delete(j, single_choice)
            dick.pop(i)
```

```python
def inrow(pist):          # check if the list of pos in pist are in the same row
    rist=[]
    for i in pist:
        rist.append(i[0])
    if len(set(rist))==1:
        return relrow(rist[0])
    return 0

def incol(pist):          # check if the list of pos in pist are in the same col
    rist=[]
    for i in pist:
        rist.append(i[1])
    if len(set(rist))==1:
        return relcol(rist[1])
    return 0

def insub(pist):           # check if the list of pos in pist are in the same subgrid
    rist=[]
    for i in pist:
        y=i[0]
        x=i[1]
        pos0=(y-y%3,x-x%3)
        rist.append(pos0)
    if len(set(rist))==1:
        return old_relsub(rist[0])
    return 0
```

```python
def nc2(nist):     # returns combinations of 2 elements from a list
    out=[]
    for i in range(len(nist)):
        for j in range(i+1,len(nist)):
            out.append((nist[i],nist[j]))
    return out
```

```python
def update2(pos, num):        # input a position and num ==> fill in the num in that pos and                   
    global dick                                          # delete that num in dick choices of related position
    global grid
    dick.pop(pos)
    for j in rel(pos):
        delete(j, num)
```

```python
def fill(pos, num):
    global dick                                          # fill that num in pos and update relation pos choices
    global grid
    dick.pop(pos)
    grid[pos[0]][pos[1]]=num
    for j in rel(pos):
        delete(j, num)  
```

```python
def guess():               # x = 0 or 1, my choice
    global dick, dick2                                          # take a guess for one of two choices in a pos
    global grid, grid2
    grid2 = copy.deepcopy(grid)            # checkpoint of grid
    dick2 = copy.deepcopy(dick)           # checkpoint of dick
    for which in (0,1):          # which choice, 0 or 1 ?
        for pos in dick:
            choices = dick.get(pos)
            if choices!=None:
                if len(choices)==2:
                    fill(pos,choices[which])
                    #print(pos, 'filled with', choices[x])
                    break
        for i in range(3):
            unique(3)
            allock(1)
            for j in range(2):
                dou()
        if len(dick)!=0:
            load()
    return grid
```

```python
def load():             # return to checkpoint
    global dick, dick2                                         
    global grid, grid2
    grid = copy.deepcopy(grid2)  
    dick = copy.deepcopy(dick2)    
```

```python
def solve(x):        # sudoku solver with combinations of all the techniques
    global grid
    grid=matint(x)
    allchoice()
    for i in range(3):
        unique(3)
        allock(1)
        for j in range(2):
            dou()
    if len(dick)!=0:
        guess()
    print(grid)
    return len(dick)
```
